```python
class KayakRentalShop:
    
    def __init__(self, totalKayaks):
        self.totalKayaks = int(totalKayaks) # Renting and Returning changes inventory amount
    
    def calculatePrice(self, amount, rate):
        if amount > 2 and amount < 5:
            discount = .3
        else:
            discount = 0
        
        cPrice = amount * rate * (1 - discount)
        if cPrice == 0:
            print("Error: Wrong Input Expected")
        
        print("Your Total is: $")
        print(cPrice)
        return cPrice
           
```

```python
class Customer:
    def __init__(self, name, numKayaks):
        self.name = name
        self.numKayaks = int(numKayaks)
        
    def trackKayaks(self): # number of kayaks per customer is counted, renting/returning changes amount
        print(self.numKayaks)
```

```python
def programRun():
    
    totalKayaks = 50 # initialize starting inventory
    shopProgram = KayakRentalShop(totalKayaks) 
    
    print("Welcome to the Kayak Rental Shop.\n") # Write Exit at any time?
    
    name = input("Enter Your Name: ") # user's name saved
    user = Customer(name, 0) 
    
    # START PROGRAM
    while True: # Program keeps running until emergency stop
        choice = int(input("\nSELECT: 1. RENT | 2. RETURN | 3. STOP PROGRAM (Enter Number) \n")) # accept num
        
        if choice == 1: # RENT Menu
            print("\n*** RENT A KAYAK MENU ***\n")
            numKayaks = int(input("How many kayaks would you like to rent? (Enter Number) "))

            while numKayaks > shopProgram.totalKayaks:
                print("Unfortunately we don't have enough Kayaks.")
                numKayaks = int(input("Please enter a smaller amount of kayaks to rent. "))

            rentTime = input("What time duration? [Hourly, Daily, Weekly] (Enter as written) ")
            if rentTime == "Hourly":
                rentRate = 10
            elif rentTime == "Daily":
                rentRate = 30
            elif rentTime == "Weekly":
                rentRate = 50
            else:
                rentRate = 0

            shopProgram.calculatePrice(numKayaks, rentRate) 
            shopProgram.totalKayaks -= numKayaks
            print("Thank You For Renting!")
            print("Kayak Rental Shop - Remaining Inventory")
            print(shopProgram.totalKayaks)
            
            user.numKayaks += numKayaks
            print(name + " - Rented Kayaks: ")
            user.trackKayaks()

        elif choice == 2: # RETURN Menu
            print("\n*** RETURN A KAYAK MENU ***\n")
            returnAmount = int(input("How many kayaks are you returning? "))

            shopProgram.totalKayaks += returnAmount
            print("Thank You for returning your kayak(s)")
            print("Kayak Rental Shop - New Inventory: ")
            print(shopProgram.totalKayaks)
            
            user.numKayaks -= returnAmount
            print(name + " - Rented Kayaks: ")
            user.trackKayaks()
        elif choice == 3:
            print("END PROGRAM")
            break
        else:
            print("Error: Please Pick Again")
```

```python
if __name__ == "__main__": # NOTE: PROGRAM DOES NOT ACCOUNT FOR WRONG INPUTS, Program intended for one user at a time
    programRun()             
    # I put the invoice in the rent menu, because it makes more sense to me to ask for price first
    # For future implementation, I would like to add a database with stored orders and details
    # Currently this program only adds new users and must be kept open to return active orders
    # Meaning there is a bug if the first action the user does is return kayaks, again - PROGRAM NEEDS CORRECT INPUTS
    # For example, I could create an order class, generate Order ID with order contents, etc.
    # one customer could have multiple orders etc.
```
